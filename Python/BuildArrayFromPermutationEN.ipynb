The QuestionProblem: Build Array from PermutationGiven a zero-based permutation nums (0-indexed), build an array ans of the same length where ans[i] = nums[nums[i]] for each $0 \le i < nums.length$ and return it.Note: A zero-based permutation is an array containing distinct integers from $0$ to $n-1$.

```python
def build_array(nums):
    # This creates a new list where each element at index i 
    # is the value found at nums[nums[i]]
    return [nums[nums[i]] for i in range(len(nums))]

# Example usage:
nums = [0, 2, 1, 5, 3, 4]
print(build_array(nums))  # Output: [0, 1, 2, 4, 5, 3]
```

3. The ExplanationThe logic here is Nested Indexing. Instead of just taking the value at the current position $i$, we use that value as a "pointer" to look up another value within the same array.Time Complexity: $O(n)$ because we visit each index exactly once.Space Complexity: $O(n)$ because we create a new array to store the results.
