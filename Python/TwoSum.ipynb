
# 5. مجموع الرقمين (Two Sum)

### **1. السؤال (The Question)**

**المشكلة:**
لديك مصفوفة من الأعداد الصحيحة `nums` ورقم صحيح يمثل الهدف `target`.
المطلوب هو إرجاع **مؤشري** (Indices) الرقمين اللذين يساوي مجموعهما قيمة الهدف `target`.

**الشروط:**

* يمكنك افتراض وجود **حل واحد فقط** لكل مدخل.
* لا يمكنك استخدام نفس العنصر مرتين (نفس الـ Index).
* يمكنك إرجاع النتيجة بأي ترتيب.

**مثال 1:**

* **الإدخال (Input):** `nums = [2, 7, 11, 15]`, `target = 9`
* **المخرجات (Output):** `[0, 1]`
* **الشرح:** لأن `nums[0] + nums[1] == 9`، نرجع `[0, 1]`.

**مثال 2:**

* **الإدخال (Input):** `nums = [3, 2, 4]`, `target = 6`
* **المخرجات (Output):** `[1, 2]`
* **الشرح:** الرقمين هما 2 و 4.

**مثال 3:**

* **الإدخال (Input):** `nums = [3, 3]`, `target = 6`
* **المخرجات (Output):** `[0, 1]`

### **2. الحل الأمثل (بلغة بايثون - Hash Map)**

سنستخدم **قاموس (Dictionary)** لتخزين الأرقام التي مررنا عليها، لنتمكن من البحث عنها بسرعة فائقة.

```python
def twoSum(nums, target):
    # إنشاء قاموس لتخزين القيمة وموقعها
    # Key: الرقم, Value: الـ Index تبعه
    prevMap = {} 
    
    for i, n in enumerate(nums):
        # ما هو الرقم الذي ينقصنا لنصل للهدف؟
        diff = target - n
        
        # هل رأينا الرقم الناقص سابقاً؟
        if diff in prevMap:
            # نعم! وجدنا الحل، نرجع موقع الرقم الناقص وموقع الرقم الحالي
            return [prevMap[diff], i]
        
        # لا، لم نجده. قم بتخزين الرقم الحالي وموقعه للبحث عنه لاحقاً
        prevMap[n] = i
        
    return [] # في حال لم يوجد حل (رغم أن السؤال يضمن وجود حل)

```

### **4. تحليل التعقيد (Complexity Analysis)**

* **التعقيد الزمني (Time Complexity):** .
نحن نمر على القائمة مرة واحدة فقط. عمليات البحث والإضافة في الـ Dictionary تأخذ وقتاً ثابتاً  تقريباً.
* **تعقيد المساحة (Space Complexity):** .
في أسوأ الأحوال، قد نضطر لتخزين كل عناصر المصفوفة في القاموس قبل أن نجد الحل في آخر رقمين.

````python
def TwoSum(arr, target):
    # إنشاء قاموس لتخزين القيمة وموقعها
    # Key: الرقم, Value: الـ Index تبعه
    prevMap = {} 
    
    for i, n in enumerate(arr):
        # ما هو الرقم الذي ينقصنا لنصل للهدف؟
        diff = target - n
        
        # هل رأينا الرقم الناقص سابقاً؟
        if diff in prevMap:
            # نعم! وجدنا الحل، نرجع موقع الرقم الناقص وموقع الرقم الحالي
            return [prevMap[diff], i]
        
        # لا، لم نجده. قم بتخزين الرقم الحالي وموقعه للبحث عنه لاحقاً
        prevMap[n] = i
    return []
````

````python
arr = [1, 2, 3, 4, 5]
target = 9
print(TwoSum(arr, target))  # Expected output: [3, 4] because arr[3] + arr[4] = 4 + 5 = 9
````
