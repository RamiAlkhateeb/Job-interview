# 4. ترتيب المصفوفة (Sort an Array)

### **1. السؤال (The Question)**

**المشكلة:**
لديك مصفوفة من الأعداد الصحيحة `nums`، والمطلوب هو ترتيب هذه المصفوفة **تصاعدياً** (من الأصغر للأكبر) وإرجاعها.

**الشروط:**

* يجب حل المشكلة بـ **تعقيد زمني قدره **.
* يجب الحل بأقل تعقيد مساحة ممكن.
* **ممنوع** استخدام دوال الترتيب الجاهزة (مثل `sort()` أو `Array.Sort()`).

**مثال 1:**

* **الإدخال (Input):** `nums = [5, 2, 3, 1]`
* **المخرجات (Output):** `[1, 2, 3, 5]`
* **الشرح:** تغيرت مواقع بعض الأرقام لتصبح مرتبة، بينما بقيت أرقام أخرى (مثل 2 و 3) في ترتيبها النسبي الصحيح.

**مثال 2:**

* **الإدخال (Input):** `nums = [5, 1, 1, 2, 0, 0]`
* **المخرجات (Output):** `[0, 0, 1, 1, 2, 5]`
* **الشرح:** القيم ليست بالضرورة فريدة (قد تتكرر).

### **2. الحل (بلغة بايثون - Merge Sort)**

سنستخدم خوارزمية **Merge Sort** لأنها تضمن دائماً الوقت المطلوب  في أسوأ الحالات وأفضلها.

```python
def sortArray(nums):
    # حالة التوقف: إذا كانت المصفوفة تحتوي عنصراً واحداً أو فارغة، فهي مرتبة أصلاً
    if len(nums) <= 1:
        return nums

    # 1. مرحلة التقسيم (Divide)
    mid = len(nums) // 2
    left_half = nums[:mid]
    right_half = nums[mid:]

    # استدعاء ذاتي (Recursion) لتقسيم النصفين حتى نصل لعنصر واحد
    sortArray(left_half)
    sortArray(right_half)

    # 2. مرحلة الدمج (Conquer / Merge)
    i = 0 # مؤشر للنصف الأيسر
    j = 0 # مؤشر للنصف الأيمن
    k = 0 # مؤشر للمصفوفة الأصلية (النتيجة)

    # مقارنة العناصر ودمجها في المصفوفة الأصلية
    while i < len(left_half) and j < len(right_half):
        if left_half[i] < right_half[j]:
            nums[k] = left_half[i]
            i += 1
        else:
            nums[k] = right_half[j]
            j += 1
        k += 1

    # إضافة ما تبقى من عناصر في النصف الأيسر (إن وجدت)
    while i < len(left_half):
        nums[k] = left_half[i]
        i += 1
        k += 1

    # إضافة ما تبقى من عناصر في النصف الأيمن (إن وجدت)
    while j < len(right_half):
        nums[k] = right_half[j]
        j += 1
        k += 1

    return nums

```

### **3. التحليل التقني: رحلة الوصول للحل (التفكير بصوت عالٍ)**

عندما يُطلب مني ترتيب مصفوفة في مقابلة، ويوضع شرط ، فهذا يلغي فوراً الحلول البسيطة. دعونا نفكر كيف نصل للحل الأمثل:

#### **المرحلة الأولى: استبعاد الحلول البطيئة**

أول ما يخطر ببالي هو **Bubble Sort** أو **Insertion Sort**.

* هذه الخوارزميات سهلة جداً (مقارنة كل رقم بجاره وتبديلهم).
* لكن مشكلتها أن تعقيدها الزمني .
* **النتيجة:** لو أعطاني مصفوفة فيها 100,000 رقم، هذه الخوارزميات ستأخذ وقتاً طويلاً جداً (Time Limit Exceeded). إذن، يجب استبعادها.

#### **المرحلة الثانية: استراتيجية "فرّق تسد" (Divide and Conquer)**

الشرط هو . هذا الرقم دائماً يلمح لشيء واحد: **شجرة ثنائية (Binary Tree) وتقسيم**.
الفكرة: ترتيب مصفوفة ضخمة أمر صعب، لكن ترتيب مصفوفة فيها عنصرين فقط أمر تافه، أليس كذلك؟

* ماذا لو قسمنا المصفوفة لنصفين؟ ثم قسمنا النصفين لأرباع؟ وهكذا حتى يصبح لدينا أجزاء صغيرة جداً (عنصر واحد).
* المصفوفة التي فيها عنصر واحد تعتبر "مرتبة" حكماً.

#### **المرحلة الثالثة: سحر "الدمج" (Merging)**

الآن التحدي الحقيقي: لدي مصفوفتان صغيرتان مرتبتان، كيف أجمعهما في مصفوفة واحدة كبيرة مرتبة؟
تخيل أن لديك كومتين من أوراق اللعب المرتبة على الطاولة، وتريد دمجهما.

1. تنظر لأعلى ورقة في الكومة اليمين وأعلى ورقة في الكومة اليسار.
2. تأخذ الأصغر بينهما وتضعها في الكومة الجديدة.
3. تكرر العملية حتى تنتهي الأوراق.
هذه العملية سريعة جداً وخطية ().

#### **الخلاصة:**

خوارزمية **Merge Sort** تقوم بالعمليتين:

1. **Divide:** تقسم المصفوفة للنصف باستمرار ( من المرات).
2. **Merge:** تدمج الأجزاء المرتبة ( من العمليات).

### **4. تحليل التعقيد (Complexity Analysis)**

* **التعقيد الزمني (Time Complexity):** .
* عملية التقسيم تخلق شجرة عمقها .
* في كل مستوى من الشجرة، نقوم بزيارة ودمج  عنصر.
* هذا هو أسرع وقت ممكن لترتيب المقارنة (Comparison-based sorting).


* **تعقيد المساحة (Space Complexity):** .
* في لغة بايثون (والحل المكتوب أعلاه)، نقوم بإنشاء مصفوفات فرعية مؤقتة (`left_half`, `right_half`) أثناء التقسيم. هذا يستهلك ذاكرة إضافية.
* *ملاحظة للمحترفين:* إذا كان القيد صارماً جداً على المساحة (أي نريد )، يجب استخدام خوارزمية **Heap Sort**، لكنها غالباً ما تكون أبطأ قليلاً في الواقع العملي وغير مستقرة (Unstable)، لذا يُفضل Merge Sort في معظم الحالات التعليمية والمقابلات ما لم يُطلب غير ذلك صراحة.

````python
def sortArray(nums):
    # حالة التوقف: إذا كانت المصفوفة تحتوي عنصراً واحداً أو فارغة، فهي مرتبة أصلاً
    if len(nums) <= 1:
        return nums

    # 1. مرحلة التقسيم (Divide)
    mid = len(nums) // 2
    left_half = nums[:mid]
    right_half = nums[mid:]

    # استدعاء ذاتي (Recursion) لتقسيم النصفين حتى نصل لعنصر واحد
    sortArray(left_half)
    sortArray(right_half)

    # 2. مرحلة الدمج (Conquer / Merge)
    i = 0 # مؤشر للنصف الأيسر
    j = 0 # مؤشر للنصف الأيمن
    k = 0 # مؤشر للمصفوفة الأصلية (النتيجة)

    # مقارنة العناصر ودمجها في المصفوفة الأصلية
    while i < len(left_half) and j < len(right_half):
        if left_half[i] < right_half[j]:
            nums[k] = left_half[i]
            i += 1
        else:
            nums[k] = right_half[j]
            j += 1
        k += 1

    # إضافة ما تبقى من عناصر في النصف الأيسر (إن وجدت)
    while i < len(left_half):
        nums[k] = left_half[i]
        i += 1
        k += 1

    # إضافة ما تبقى من عناصر في النصف الأيمن (إن وجدت)
    while j < len(right_half):
        nums[k] = right_half[j]
        j += 1
        k += 1

    return nums
````
