
#  تسلق السلالم (Climbing Stairs)

### **1. السؤال**

**المشكلة:** أنت تتسلق سلماً يتكون من  من الدرجات. في كل مرة، يمكنك إما صعود **درجة واحدة أو درجتين**. المطلوب هو إيجاد عدد الطرق الفريدة والمختلفة التي يمكنك من خلالها الوصول إلى القمة.

### **2. الحل (بلغة بايثون - المحسن)**

```python
def climbStairs(n):
    # الحالات الأساسية
    if n <= 2:
        return n
    
    # تحسين المساحة: نحتاج فقط لآخر قيمتين وليس لمصفوفة كاملة
    prev1 = 2 # تمثل n-1
    prev2 = 1 # تمثل n-2
    
    for i in range(3, n + 1):
        # القيمة الحالية هي مجموع الطريقتين السابقتين
        current = prev1 + prev2
        # تحديث القيم للخطوة القادمة
        prev2 = prev1
        prev1 = current
        
    return prev1

```

### **4. تحليل التعقيد (Complexity Analysis)**

* **التعقيد الزمني (Time Complexity):** هو  لأننا نقوم بعمل دورة واحدة (Loop) تبدأ من الدرجة 3 وصولاً إلى . كل عملية داخل الدورة هي عملية حسابية بسيطة ().
* **تعقيد المساحة (Space Complexity):** هو . بفضل استخدامنا لمتغيرين فقط (`prev1`, `prev2`) بدلاً من مصفوفة كاملة، استطعنا الحفاظ على ثبات استهلاك الذاكرة مهما كبر عدد الدرجات.


````python
def climbStairs(n):
    # الحالات الأساسية
    if n <= 2:
        return n
    
    # تحسين المساحة: نحتاج فقط لآخر قيمتين وليس لمصفوفة كاملة
    prev1 = 2 # تمثل n-1
    prev2 = 1 # تمثل n-2
    
    for i in range(3, n + 1):
        # القيمة الحالية هي مجموع الطريقتين السابقتين
        current = prev1 + prev2
        # تحديث القيم للخطوة القادمة
        prev2 = prev1
        prev1 = current
        
    return prev1
````
